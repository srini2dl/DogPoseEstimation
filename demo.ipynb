```python
import matplotlib.pyplot as plt
import numpy as np
import cv2
import pandas as pd
from skimage import io
import Constants

# Opens the Video file
cap = cv2.VideoCapture('videos/test.mp4')
i = 0
checkFrame = 20
while (cap.isOpened()):
    ret, frame = cap.read()
    if ret == False:
        break
    if i == checkFrame:
        imagePath = 'videos/sample_frame' + str(i) + '.jpg'
        cv2.imwrite(imagePath, frame)
        break
    i += 1

cap.release()
cv2.destroyAllWindows()
hdfPath = 'videos/testshuffle1_7500.h5'
predictedDataFrame = pd.read_hdf(hdfPath, 'df_with_missing')
im=io.imread(imagePath)
h,w,numcolors=np.shape(im)
dotsize = 20
```

```python
hdfPath = 'videos/testshuffle1_7500.h5'
predictedDataFrame = pd.read_hdf(hdfPath, 'df_with_missing')
im=io.imread(imagePath)
h,w,numcolors=np.shape(im)
dotsize = 20
plt.figure(frameon=False,figsize=(w*1./60,h*1./60))
plt.subplots_adjust(left=0, bottom=0, right=1, top=1, wspace=0, hspace=0)
plt.imshow(im)
for bp in Constants.bodyparts:
    if predictedDataFrame['shuffle1_7500'][bp]['likelihood'].values[checkFrame] > 0.5:
        x = predictedDataFrame['shuffle1_7500'][bp]['x'].values[checkFrame]
        y = predictedDataFrame['shuffle1_7500'][bp]['y'].values[checkFrame]
        plt.plot(x, y, '.', ms=dotsize, alpha=0.7)

plt.show()
```

```python
im=io.imread(imagePath)
h,w,numcolors=np.shape(im)
dotsize = 20
plt.figure(frameon=False,figsize=(w*1./60,h*1./60))
plt.subplots_adjust(left=0, bottom=0, right=1, top=1, wspace=0, hspace=0)
plt.imshow(im, alpha=0.3)

for bone in Constants.bones:
    jointXCorordinates = []
    jointYCorordinates = []
    for joints in bone:
        if predictedDataFrame['shuffle1_7500'][joints[0]]['likelihood'].values[checkFrame] > 0.5 and predictedDataFrame['shuffle1_7500'][joints[1]]['likelihood'].values[checkFrame] > 0.6 :
            x = predictedDataFrame['shuffle1_7500'][joints[0]]['x'].values[checkFrame]
            y = predictedDataFrame['shuffle1_7500'][joints[0]]['y'].values[checkFrame]
            x2 = predictedDataFrame['shuffle1_7500'][joints[1]]['x'].values[checkFrame]
            y2 = predictedDataFrame['shuffle1_7500'][joints[1]]['y'].values[checkFrame]
            jointXCorordinates.append(x)
            jointXCorordinates.append(x2)
            jointYCorordinates.append(y)
            jointYCorordinates.append(y2)
            plt.plot(x, y, '.', ms=dotsize, alpha=0.7)
            plt.plot(x2, y2, '.', ms=dotsize, alpha=0.7)
    plt.plot(jointXCorordinates, jointYCorordinates)        
```
